```python
from scipy.optimize import linprog
```

```python
# Coeficientes de la función que serán minimizados
c = [-150, -175]

# Coeficientes de las desigualdades de restricciones
A_ub = [[7, 11], [10, 8], [1, 0], [0, 1],]

# Elementos del vector b
b_ub = [77, 80, 9, 6]

# Rango de los valores x1 y x2
x1_bounds = (0, None)
x2_bounds = (0, None)
bounds = [x1_bounds, x2_bounds]

# Resultado
result_prueba = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
```

```python
result_prueba
```
